```python
import pandas as pd

def load_data(filepath):
    """Load CSV data and parse dates."""
    return pd.read_csv(filepath, parse_dates=['MonthData'])

def preprocess_data(df):
    """Preprocess the DataFrame by extracting year and month."""
    df['MonthData'] = pd.to_datetime(df['MonthData'])
    df['Year'] = df['MonthData'].dt.year
    df['Month'] = df['MonthData'].dt.month
    return df

def transform_amount(group):
    """Apply transformation on Amount column for IS report type."""
    group = group.sort_values(by=['Year', 'Month'])
    group['Amount'] = group['Amount'].diff().fillna(group['Amount'])
    return group

def split_report_types(df):
    """Separate IS and BS report types and apply transformation."""
    df_is = df[df['ReportType'] == 'IS'].groupby(['unique_id', 'Year'], group_keys=False).apply(transform_amount)
    df_bs = df[df['ReportType'] == 'BS']
    return pd.concat([df_is, df_bs], ignore_index=True)

def rename_columns(df):
    """Rename columns to match expected format."""
    df = df[['MonthData', 'Amount', 'unique_id']]
    return df.rename(columns={'MonthData': 'ds', 'Amount': 'y'})

def split_train_valid(df, forecast_horizon=6):
    """Split DataFrame into train and validation sets."""
    train, valid = pd.DataFrame(columns=df.columns), pd.DataFrame(columns=df.columns)
    
    for unique in df['unique_id'].unique():
        temp_df = df[df['unique_id'] == unique]
        train_temp = temp_df.iloc[:-forecast_horizon]
        valid_temp = temp_df.iloc[-forecast_horizon:]
        train = pd.concat([train, train_temp])
        valid = pd.concat([valid, valid_temp])
    
    return train, valid

# Execution workflow
data_path = '5_Feb_ALL_JAN_Train (1).csv'
df = load_data(data_path)
# df = preprocess_data(df)
# df = split_report_types(df)
# df = rename_columns(df)
# train, valid = split_train_valid(df)

# print("Train DataFrame:", train.shape)
# print("Valid DataFrame:", valid.shape)
```

```python
import os
import mlflow
import boto3
import pandas as pd
import numpy as np
from prophet import Prophet
from prophet.serialize import model_to_json
from joblib import Parallel, delayed
from pathlib import Path

# AWS S3 Configuration
S3_BUCKET_NAME = "bankers-dashboard-harshit-t476902"
S3_FOLDER = "mlflow"

# MLflow Configuration
MLFLOW_TRACKING_URI = "arn:aws:sagemaker:us-east-1:590184044598:mlflow-tracking-server/bankers-dashboard-harshit-t476902"  # Replace with your MLflow server URL
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)
mlflow.set_experiment("Prophet Forecasting")

# Initialize S3 Client
s3_client = boto3.client('s3')

def save_model_to_s3_json(model, unique_id):
    """Saves Prophet model as JSON and uploads to S3."""
    model_filename = f"{unique_id}_prophet_model.json"
    local_path = Path(f"/tmp/{model_filename}")  
    s3_path = f"{S3_FOLDER}/{model_filename}"  

    with local_path.open('w') as fout:
        fout.write(model_to_json(model))

    s3_client.upload_file(str(local_path), S3_BUCKET_NAME, s3_path)
    return f"s3://{S3_BUCKET_NAME}/{s3_path}"

def run_one(unique_id):
    """Trains Prophet model for a given unique_id and logs results."""
    train_ = train[train['unique_id'] == unique_id]
    valid_ = valid[valid['unique_id'] == unique_id]

    print(f"Training unique_id: {unique_id}")
    m = Prophet()
    m.fit(train_)

    # Make Predictions
    future = m.make_future_dataframe(periods=44, freq='M', include_history=True)
    future = future.merge(valid_[['ds']], on='ds', how='left')
    forecast = m.predict(future)
    forecast['unique_id'] = unique_id

    return unique_id, m, forecast

# Run in parallel using threads to avoid PicklingError
results = Parallel(n_jobs=-1, prefer="threads")(
    delayed(run_one)(uid) for uid in train['unique_id'].unique()
)

# Post-processing: Save models, log to MLflow & S3
all_forecasts = []
for unique_id, model, forecast in results:
    # Save Model to S3
    s3_model_path = save_model_to_s3_json(model, unique_id)

    # Compute WMAPE
    merged = forecast.merge(valid, on=['ds', 'unique_id'], how='left').dropna()
    y_true, y_pred = merged['y'], merged['yhat']
    wmape_score = np.abs(y_true - y_pred).sum() / np.abs(y_true).sum()

    # Log Model & Metrics to MLflow
    with mlflow.start_run(run_name=f"prophet_{unique_id}"):
        mlflow.log_param("unique_id", unique_id)
        mlflow.log_metric("wmape", wmape_score)
        mlflow.log_artifact(f"/tmp/{unique_id}_prophet_model.json", artifact_path=f"models/{unique_id}")

    print(f" Model {unique_id} saved to S3: {s3_model_path}, WMAPE: {wmape_score:.4f}")
    all_forecasts.append(forecast)

# Save final results
final_df = pd.concat(all_forecasts, ignore_index=True)
final_df.to_csv("Prophet_results_new.csv", index=False)
```

```python
import boto3
import pandas as pd
import os
from prophet import Prophet
from prophet.serialize import model_from_json
from io import StringIO

# AWS S3 Configuration
S3_BUCKET_NAME = "bankers-dashboard-harshit-t476902"
S3_MODEL_PATH = "mlflow/{unique_id}_prophet_model.json"  # Path template

# Initialize S3 Client
s3_client = boto3.client('s3')

def load_model_from_s3(unique_id):
    """Downloads and loads a Prophet model from S3."""
    model_s3_path = S3_MODEL_PATH.format(unique_id=unique_id)
    print(model_s3_path)
    
    try:
        # Download model content from S3
        obj = s3_client.get_object(Bucket=S3_BUCKET_NAME, Key=model_s3_path)
        model_json = obj['Body'].read().decode('utf-8')

        # Load the Prophet model
        model = model_from_json(model_json)
        print(f" Loaded model for unique_id: {unique_id} from S3")
        return model

    except Exception as e:
        print(f" Error loading model from S3: {e}")
        return None

def predict_future(unique_id, periods=6):
    """Loads model, generates future dates, and makes predictions."""
    model = load_model_from_s3(unique_id)
    if model is None:
        return None

    # Generate future dates
    future = model.make_future_dataframe(periods=periods, freq='M')

    # Predict future values
    forecast = model.predict(future)
    print(forecast.columns)

    return forecast[['ds', 'yhat']]

# Example usage
unique_id_to_predict = "10643_0_125000000"
predictions = predict_future(unique_id_to_predict, periods=44)

if predictions is not None:
    print(predictions)

#s3://bankers-dashboard-harshit-t476902/mlflow/10643_0_125000000_prophet_model.json
```
