```python
import pandas as pd
import json

def create_json_format(df):
    # Convert MonthData to datetime first
    df['MonthData'] = pd.to_datetime(df['MonthData'])
    
    output = []
    # Get unique IDs
    unique_ids = df['unique_id'].unique()
    
    # Iterate over each unique_id
    for idx, uid in enumerate(unique_ids):
        # Filter the DataFrame for the current unique_id
        df_uid = df[df['unique_id'] == uid]
        df_uid = df_uid.sort_values(by=['MonthData'])
        print(uid)

        # Create dictionary for the current unique_id
        uid_dict = {
            "start": df_uid['MonthData'].min().strftime('%Y-%m-%d'),
            "target": df_uid['Amount'].tolist(),
            "cat": df_uid['unique_id_le'].tolist()[0],
        }
        output.append(uid_dict)
    return output
```

```python
import pandas as pd
def train_test_split_by_unique_id(df, test_size=6):
    train_list = []
    test_list = []
    
    for unique_id, group in df.groupby('unique_id'):
        # Sort group by time if it's not already sorted
        group = group.sort_values(by=['MonthData'])  # Or sort by time column if you have one
        
        # Split into train and test
        train_data = group.iloc[:-test_size]
        test_data = group.iloc[-test_size:]
        
        # Append results to lists
        train_list.append(train_data)
        test_list.append(test_data)
    
    # Concatenate lists into DataFrames
    train_df = pd.concat(train_list)
    test_df = pd.concat(test_list)
    
    return train_df, test_df
```

```python
import json 

# correct np narray
class NumpyEncoder(json.JSONEncoder):
    """ Special json encoder for numpy types """
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        return json.JSONEncoder.default(self, obj)

# convert to json lines: each line is a json object
def write_dicts_to_file(path, data):
    with open(path, 'wb') as fp:
        for d in data:
            fp.write(json.dumps(d, cls=NumpyEncoder).encode("utf-8"))
            fp.write("\n".encode('utf-8'))
            
def write_dicts_to_json(path, data):
    with open(path, "wb") as file_path:
        for ts in data:
            file_path.write(json.dumps(ts).encode("utf-8"))
            file_path.write("\n".encode("utf-8"))
```

```python
import pandas as pd
import numpy as np
import json
from datetime import datetime

def evaluate_model_performance(test_df, predictions, key_value_dict, 
                             start_date='2024-05-01', end_date='2024-10-01'):
    """
    Evaluate model performance and create detailed comparison with date filtering
    and duplicate removal
    """
    # Parse predictions
    pred_data = json.loads(predictions)
    
    # Create comparison dataframe
    comparison_rows = []
    metrics_by_series = []
    
    # Convert dates to datetime and extract month start dates
    start_dt = pd.to_datetime(start_date).normalize().replace(day=1)
    end_dt = pd.to_datetime(end_date).normalize().replace(day=1)
    
    print(f"Processing predictions for period: {start_dt} to {end_dt}")
    
    # Process each time series
    seen_combinations = set()  # Track unique combinations of date and ID
    
    for idx, series_pred in enumerate(pred_data["predictions"]):
        series_id = key_value_dict[idx]
        series_actuals = test_df[test_df['unique_id'] == series_id].copy()
        
        if isinstance(series_actuals.get('MonthData'), pd.Series):
            series_actuals['MonthData'] = pd.to_datetime(series_actuals['MonthData'])
        
        # Get dates for the predictions
        try:
            start_date = series_actuals['MonthData'].min() if not series_actuals.empty else start_dt
            dates = pd.date_range(start=start_date, periods=len(series_pred['mean']), freq='MS')
        except Exception as e:
            print(f"Error with dates for series {series_id}: {e}")
            continue
        
        # Create row for each prediction point within date range
        for i in range(len(dates)):
            date = dates[i]
            
            # Skip if outside desired date range
            if date < start_dt or date > end_dt:
                continue
                
            # Create unique key for deduplication
            unique_key = f"{series_id}_{date.strftime('%Y-%m-%d')}"
            if unique_key in seen_combinations:
                continue
            seen_combinations.add(unique_key)
            
            # Find the actual value
            actual_value = None
            if not series_actuals.empty:
                matching_actuals = series_actuals[
                    series_actuals['MonthData'].dt.to_period('M') == date.to_period('M')
                ]
                if not matching_actuals.empty:
                    actual_value = matching_actuals.iloc[0].get('Amount')
            
            row = {
                'MonthData': date,
                'unique_id': series_id,
                'Actual': actual_value,
                'Predicted_Mean': float(series_pred['mean'][i]),
                'Predicted_10': float(series_pred['quantiles']['0.1'][i]),
                'Predicted_50': float(series_pred['quantiles']['0.5'][i]),
                'Predicted_90': float(series_pred['quantiles']['0.9'][i])
            }
            comparison_rows.append(row)
    
    print(f"Generated {len(comparison_rows)} comparison rows")
    
    # Create comparison DataFrame
    comparison_df = pd.DataFrame(comparison_rows)
    
    if not comparison_df.empty:
        # Remove any remaining duplicates
        comparison_df = comparison_df.drop_duplicates(
            subset=['MonthData', 'unique_id']
        ).copy()
        
        # Add identifiers based on the correct format of unique_id (DataSourceId_BankId_LineEntityId)
        try:
            # Split unique_id using underscore as separator
            comparison_df[['DataSourceId', 'BankId', 'LineEntityId']] = \
                comparison_df['unique_id'].str.split('_', expand=True)
        except Exception as e:
            print(f"Error splitting unique_id: {e}")
            # If splitting fails, try to extract components individually
            try:
                # Extract components with a more flexible approach
                comparison_df['DataSourceId'] = comparison_df['unique_id'].str.extract(r'(\d+)_')
                comparison_df['BankId'] = comparison_df['unique_id'].str.extract(r'_(\d+)_')
                comparison_df['LineEntityId'] = comparison_df['unique_id'].str.extract(r'_(\d+)$')
            except Exception as e2:
                print(f"Alternative ID extraction also failed: {e2}")
        
        # Sort by date and unique_id
        comparison_df = comparison_df.sort_values(['MonthData', 'unique_id'])
        
        # Calculate metrics by series
        for series_id in comparison_df['unique_id'].unique():
            series_data = comparison_df[comparison_df['unique_id'] == series_id]
            
            # Convert to numeric, replacing non-numeric values with NaN
            actuals = pd.to_numeric(series_data['Actual'], errors='coerce')
            predictions = pd.to_numeric(series_data['Predicted_Mean'], errors='coerce')
            
            # Filter out NaN values
            mask = ~(actuals.isna() | predictions.isna())
            valid_actuals = actuals[mask].values
            valid_predictions = predictions[mask].values
            
            if len(valid_actuals) > 0:
                # Calculate metrics
                with np.errstate(divide='ignore', invalid='ignore'):
                    mape = np.mean(np.abs((valid_actuals - valid_predictions) / valid_actuals)) * 100
                    mae = np.mean(np.abs(valid_actuals - valid_predictions))
                    rmse = np.sqrt(np.mean((valid_actuals - valid_predictions) ** 2))
                    
                    metrics_by_series.append({
                        'unique_id': series_id,
                        'MAPE': mape if not np.isinf(mape) else None,
                        'MAE': mae,
                        'RMSE': rmse
                    })
    
    # Create metrics DataFrame
    metrics_df = pd.DataFrame(metrics_by_series)
    if not metrics_df.empty:
        try:
            # Split unique_id using underscore as separator for metrics dataframe
            metrics_df[['DataSourceId', 'BankId', 'LineEntityId']] = \
                metrics_df['unique_id'].str.split('_', expand=True)
            
            # Add ReportType and LineEntity_name from test_df if available
            if 'ReportType' in test_df.columns:
                # Create a mapping of unique_id to ReportType
                report_type_map = test_df.drop_duplicates('unique_id').set_index('unique_id')['ReportType'].to_dict()
                metrics_df['ReportType'] = metrics_df['unique_id'].map(report_type_map)
            
            # Try to extract LineEntityId_name from test_df if available
            if 'LineEntityId_name' in test_df.columns:
                line_name_map = test_df.drop_duplicates('unique_id').set_index('unique_id')['LineEntityId_name'].to_dict()
                metrics_df['LineEntityId_name'] = metrics_df['unique_id'].map(line_name_map)
                
        except Exception as e:
            print(f"Error processing metrics dataframe: {e}")
            # If splitting fails, print the first few unique_ids to help debug
            print("Sample unique_ids in metrics_df:")
            print(metrics_df['unique_id'].head().to_list())
        
        # Calculate overall metrics (excluding None/inf values)
        overall_metrics = {
            'Overall_MAPE': metrics_df['MAPE'].replace([np.inf, -np.inf], np.nan).mean(),
            'Overall_MAE': metrics_df['MAE'].mean(),
            'Overall_RMSE': metrics_df['RMSE'].mean()
        }
        
        print("\nOverall Model Performance:")
        for metric, value in overall_metrics.items():
            if pd.notnull(value):
                print(f"{metric}: {value:.2f}")
            else:
                print(f"{metric}: N/A")
    
    # Save results with date range in filename
    date_suffix = f"{start_dt.strftime('%Y%m')}_to_{end_dt.strftime('%Y%m')}"
    comparison_df.to_csv(f'actuals_vs_predicted_detailed_{date_suffix}.csv', 
                        index=False)
    metrics_df.to_csv(f'model_performance_by_series_{date_suffix}.csv', 
                     index=False)
    
    return comparison_df, metrics_df
```

```python
import pandas as pd
import numpy as np
import json
from datetime import datetime

def evaluate_model_performance(test_df, predictions, key_value_dict, 
                             start_date='2024-05-01', end_date='2024-10-01'):
    """
    Evaluate model performance and create detailed comparison grouped by unique_id
    showing monthly progression for each ID with improved handling of date ranges
    and duplicate removal
    """
    # Parse predictions
    pred_data = json.loads(predictions)
    
    # Create comparison dataframe
    comparison_rows = []
    metrics_by_series = []
    
    # Convert dates to datetime and extract month start dates
    start_dt = pd.to_datetime(start_date).normalize().replace(day=1)
    end_dt = pd.to_datetime(end_date).normalize().replace(day=1)
    
    print(f"Processing predictions for period: {start_dt} to {end_dt}")
    
    # Track unique combinations of date and ID for deduplication
    seen_combinations = set()
    
    # First, collect all unique_ids
    unique_ids = sorted(key_value_dict.values())
    
    # Process each unique_id
    for unique_id in unique_ids:
        print(f"Processing unique_id: {unique_id}")
        
        # Find the corresponding series in predictions
        series_idx = None
        for idx, series_id in key_value_dict.items():
            if series_id == unique_id:
                series_idx = idx
                break
                
        if series_idx is None:
            print(f"Warning: No predictions found for {unique_id}")
            continue
            
        series_pred = pred_data["predictions"][series_idx]
        series_actuals = test_df[test_df['unique_id'] == unique_id].copy()
        
        if isinstance(series_actuals.get('MonthData'), pd.Series):
            series_actuals['MonthData'] = pd.to_datetime(series_actuals['MonthData'])
        
        # Get dates for the predictions
        try:
            start_date_series = series_actuals['MonthData'].min() if not series_actuals.empty else start_dt
            dates = pd.date_range(start=start_date_series, periods=len(series_pred['mean']), freq='MS')
        except Exception as e:
            print(f"Error with dates for series {unique_id}: {e}")
            # Fallback to predefined date range
            dates = pd.date_range(start=start_dt, end=end_dt, freq='MS')
        
        # Create row for each prediction point within date range
        for i, date in enumerate(dates):
            # Skip if outside desired date range
            if date < start_dt or date > end_dt:
                continue
                
            # Create unique key for deduplication
            unique_key = f"{unique_id}_{date.strftime('%Y-%m-%d')}"
            if unique_key in seen_combinations:
                continue
            seen_combinations.add(unique_key)
            
            # Find the actual value
            actual_value = None
            if not series_actuals.empty:
                matching_actuals = series_actuals[
                    series_actuals['MonthData'].dt.to_period('M') == date.to_period('M')
                ]
                if not matching_actuals.empty:
                    actual_value = matching_actuals.iloc[0].get('Amount')
            
            # Only add row if we have the prediction for this index
            if i < len(series_pred['mean']):
                row = {
                    'MonthData': date,
                    'unique_id': unique_id,
                    'Actual': actual_value,
                    'Predicted_Mean': float(series_pred['mean'][i]),
                    'Predicted_10': float(series_pred['quantiles']['0.1'][i]),
                    'Predicted_50': float(series_pred['quantiles']['0.5'][i]),
                    'Predicted_90': float(series_pred['quantiles']['0.9'][i])
                }
                comparison_rows.append(row)
    
    print(f"Generated {len(comparison_rows)} comparison rows")
    
    # Create comparison DataFrame
    comparison_df = pd.DataFrame(comparison_rows)
    
    if not comparison_df.empty:
        # Remove any remaining duplicates
        comparison_df = comparison_df.drop_duplicates(
            subset=['MonthData', 'unique_id']
        ).copy()
        
        # Add identifiers based on the format of unique_id
        try:
            # First try the new format with ** separators
            comparison_df[['DataSourceId', 'ReportType', 'LineEntityId', 'BankName', 
                          'LineEntityId_name', 'BankId']] = \
                comparison_df['unique_id'].str.split(r'\*\*', expand=True)
        except Exception as e:
            print(f"Error splitting unique_id with ** separator: {e}")
            # Fall back to the old format with _ separators
            try:
                comparison_df[['DataSourceId', 'BankId', 'LineEntityId']] = \
                    comparison_df['unique_id'].str.split('_', expand=True)
            except Exception as e2:
                print(f"Error splitting unique_id with _ separator: {e2}")
                # If splitting fails, try to extract components individually
                try:
                    # Extract components with a more flexible approach
                    comparison_df['DataSourceId'] = comparison_df['unique_id'].str.extract(r'(\d+)_')
                    comparison_df['BankId'] = comparison_df['unique_id'].str.extract(r'_(\d+)_')
                    comparison_df['LineEntityId'] = comparison_df['unique_id'].str.extract(r'_(\d+)$')
                except Exception as e3:
                    print(f"Alternative ID extraction also failed: {e3}")
        
        # Sort first by unique_id, then by date
        comparison_df = comparison_df.sort_values(['unique_id', 'MonthData'])
        
        # Calculate metrics by series
        for unique_id in comparison_df['unique_id'].unique():
            series_data = comparison_df[comparison_df['unique_id'] == unique_id]
            
            # Convert to numeric, replacing non-numeric values with NaN
            actuals = pd.to_numeric(series_data['Actual'], errors='coerce')
            predictions = pd.to_numeric(series_data['Predicted_Mean'], errors='coerce')
            
            # Filter out NaN values
            mask = ~(actuals.isna() | predictions.isna())
            valid_actuals = actuals[mask].values
            valid_predictions = predictions[mask].values
            
            if len(valid_actuals) > 0:
                # Calculate metrics
                with np.errstate(divide='ignore', invalid='ignore'):
                    mape = np.mean(np.abs((valid_actuals - valid_predictions) / valid_actuals)) * 100
                    mae = np.mean(np.abs(valid_actuals - valid_predictions))
                    rmse = np.sqrt(np.mean((valid_actuals - valid_predictions) ** 2))
                    
                    metrics_by_series.append({
                        'unique_id': unique_id,
                        'MAPE': mape if not np.isinf(mape) else None,
                        'MAE': mae,
                        'RMSE': rmse
                    })
    
    # Create metrics DataFrame
    metrics_df = pd.DataFrame(metrics_by_series)
    if not metrics_df.empty:
        # Add identifier columns to metrics_df based on format of unique_id
        try:
            # First try the new format with ** separators
            metrics_df[['DataSourceId', 'ReportType', 'LineEntityId', 'BankName', 
                       'LineEntityId_name', 'BankId']] = \
                metrics_df['unique_id'].str.split(r'\*\*', expand=True)
        except Exception as e:
            print(f"Error splitting metrics unique_id with ** separator: {e}")
            # Fall back to the old format with _ separators
            try:
                metrics_df[['DataSourceId', 'BankId', 'LineEntityId']] = \
                    metrics_df['unique_id'].str.split('_', expand=True)
                
                # Add ReportType and LineEntity_name from test_df if available
                if 'ReportType' in test_df.columns:
                    # Create a mapping of unique_id to ReportType
                    report_type_map = test_df.drop_duplicates('unique_id').set_index('unique_id')['ReportType'].to_dict()
                    metrics_df['ReportType'] = metrics_df['unique_id'].map(report_type_map)
                
                # Try to extract LineEntityId_name from test_df if available
                if 'LineEntityId_name' in test_df.columns:
                    line_name_map = test_df.drop_duplicates('unique_id').set_index('unique_id')['LineEntityId_name'].to_dict()
                    metrics_df['LineEntityId_name'] = metrics_df['unique_id'].map(line_name_map)
            except Exception as e2:
                print(f"Error processing metrics dataframe: {e2}")
                # If splitting fails, print the first few unique_ids to help debug
                print("Sample unique_ids in metrics_df:")
                print(metrics_df['unique_id'].head().to_list())
        
        # Calculate overall metrics (excluding None/inf values)
        overall_metrics = {
            'Overall_MAPE': metrics_df['MAPE'].replace([np.inf, -np.inf], np.nan).mean(),
            'Overall_MAE': metrics_df['MAE'].mean(),
            'Overall_RMSE': metrics_df['RMSE'].mean()
        }
        
        print("\nOverall Model Performance:")
        for metric, value in overall_metrics.items():
            if pd.notnull(value):
                print(f"{metric}: {value:.2f}")
            else:
                print(f"{metric}: N/A")
    
    # Save results with date range in filename
    date_suffix = f"{start_dt.strftime('%Y%m')}_to_{end_dt.strftime('%Y%m')}"
    comparison_df.to_csv(f'actuals_vs_predicted_by_id_{date_suffix}.csv', index=False)
    metrics_df.to_csv(f'model_performance_by_id_{date_suffix}.csv', index=False)
    
    return comparison_df, metrics_df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def generate_comprehensive_report(comparison_df, metrics_df, output_filename='comprehensive_analysis.pdf'):
    """
    Generate a single comprehensive PDF report for time series analysis
    
    Modified to work with the existing data structure that uses BankId and LineEntityId
    instead of BankName and LineEntityId_name.
    
    Parameters:
    -----------
    comparison_df : pandas DataFrame
        DataFrame containing actual vs predicted values with columns for MonthData, 
        unique_id, Actual, Predicted_Mean, DataSourceId, BankId, LineEntityId
    
    metrics_df : pandas DataFrame
        DataFrame containing performance metrics with columns for unique_id,
        MAPE, MAE, RMSE, DataSourceId, BankId, LineEntityId
        
    output_filename : str
        Filename for the output PDF report
    """
    
    # Convert data types to ensure proper handling
    comparison_df['MonthData'] = pd.to_datetime(comparison_df['MonthData'])
    comparison_df['Actual'] = pd.to_numeric(comparison_df['Actual'], errors='coerce')
    comparison_df['Predicted_Mean'] = pd.to_numeric(comparison_df['Predicted_Mean'], errors='coerce')
    metrics_df['MAPE'] = pd.to_numeric(metrics_df['MAPE'], errors='coerce')
    metrics_df['MAE'] = pd.to_numeric(metrics_df['MAE'], errors='coerce')
    metrics_df['RMSE'] = pd.to_numeric(metrics_df['RMSE'], errors='coerce')
    
    with PdfPages(output_filename) as pdf:
        # Title page
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.7, 'Comprehensive Time Series Analysis Report', 
                ha='center', va='center', fontsize=24, fontweight='bold')
        plt.text(0.5, 0.5, f'Generated on {pd.Timestamp.now().strftime("%Y-%m-%d %H:%M:%S")}',
                ha='center', va='center', fontsize=14)
        summary_stats = (
            f"Total Banks: {len(comparison_df.BankId.unique())}\n"
            f"Total Line Items: {len(comparison_df.LineEntityId.unique())}\n"
            f"Date Range: {comparison_df['MonthData'].min().strftime('%Y-%m-%d')} to {comparison_df['MonthData'].max().strftime('%Y-%m-%d')}"
        )
        plt.text(0.5, 0.3, summary_stats, ha='center', va='center', fontsize=12)
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        # Section 1: Individual Bank Analysis
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.5, 'Section 1: Individual Bank Analysis', 
                ha='center', va='center', fontsize=20, fontweight='bold')
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        for bank_id in sorted(comparison_df['BankId'].unique()):
            print(f"Processing bank: {bank_id}")
            df_subset = comparison_df[comparison_df['BankId'] == bank_id].copy()
            
            # Bank summary plot
            fig = plt.figure(figsize=(15, 10))
            plt.suptitle(f'Analysis for Bank ID: {bank_id}', fontsize=16, y=0.95)
            
            # Time series plot
            plt.subplot(2, 1, 1)
            for line_id in df_subset['LineEntityId'].unique():
                line_data = df_subset[df_subset['LineEntityId'] == line_id]
                plt.plot(line_data['MonthData'], line_data['Actual'], 
                        label=f'Line {line_id} (Actual)', linestyle='-', marker='o')
                plt.plot(line_data['MonthData'], line_data['Predicted_Mean'], 
                        label=f'Line {line_id} (Predicted)', linestyle='--', marker='x')
            
            plt.title('Time Series by Line Item')
            plt.xticks(rotation=45)
            plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.grid(True, alpha=0.3)
            
            # Error analysis
            plt.subplot(2, 1, 2)
            errors = df_subset['Predicted_Mean'] - df_subset['Actual']
            plt.bar(df_subset['MonthData'], errors, alpha=0.6)
            plt.title('Prediction Errors')
            plt.xticks(rotation=45)
            plt.grid(True, alpha=0.3)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close()
        
        # Section 2: Line Item Analysis
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.5, 'Section 2: Line Item Analysis', 
                ha='center', va='center', fontsize=20, fontweight='bold')
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        # Limit to top 10 line items by volume if there are many
        line_items = sorted(comparison_df['LineEntityId'].unique())
        if len(line_items) > 10:
            # Find top 10 line items by total volume
            line_volumes = comparison_df.groupby('LineEntityId')['Actual'].sum().abs()
            top_lines = line_volumes.nlargest(10).index.tolist()
            print(f"Analyzing top 10 line items out of {len(line_items)} total")
            line_items = top_lines
        
        for line_id in line_items:
            print(f"Processing line item: {line_id}")
            df_subset = comparison_df[comparison_df['LineEntityId'] == line_id].copy()
            
            # Line item summary plot
            fig = plt.figure(figsize=(15, 10))
            plt.suptitle(f'Analysis for Line ID: {line_id}', fontsize=16, y=0.95)
            
            # Box plot of actuals vs predictions
            plt.subplot(2, 1, 1)
            data_to_plot = [df_subset['Actual'].dropna(), df_subset['Predicted_Mean'].dropna()]
            plt.boxplot(data_to_plot, labels=['Actual', 'Predicted'])
            plt.title('Distribution of Values')
            plt.grid(True, alpha=0.3)
            
            # Time series by bank
            plt.subplot(2, 1, 2)
            for bank_id in df_subset['BankId'].unique():
                bank_data = df_subset[df_subset['BankId'] == bank_id]
                plt.plot(bank_data['MonthData'], bank_data['Actual'], 
                        label=f'Bank {bank_id} (Actual)', marker='o')
            plt.title('Time Series by Bank')
            plt.xticks(rotation=45)
            plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.grid(True, alpha=0.3)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close()
        
        # Overall Performance Metrics
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.5, 'Section 3: Overall Performance Metrics', 
                ha='center', va='center', fontsize=20, fontweight='bold')
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        # Overall metrics plot
        fig = plt.figure(figsize=(15, 10))
        plt.suptitle('Overall Performance Metrics', fontsize=16, y=0.95)
        
        # MAPE by bank - with error handling for infinite/NaN values
        plt.subplot(2, 1, 1)
        mape_by_bank = metrics_df.groupby('BankId')['MAPE'].mean()
        mape_by_bank = mape_by_bank.replace([np.inf, -np.inf], np.nan).dropna().sort_values()
        
        if not mape_by_bank.empty:
            plt.barh(range(len(mape_by_bank)), mape_by_bank.values)
            plt.yticks(range(len(mape_by_bank)), mape_by_bank.index)
            plt.title('Average MAPE by Bank')
            plt.xlabel('MAPE (%)')
        else:
            plt.text(0.5, 0.5, 'No valid MAPE data available', ha='center')
            plt.title('Average MAPE by Bank')
        
        # RMSE by line item
        plt.subplot(2, 1, 2)
        rmse_by_line = metrics_df.groupby('LineEntityId')['RMSE'].mean().sort_values()
        
        if not rmse_by_line.empty:
            if len(rmse_by_line) > 15:  # If too many line items, show top 15
                rmse_by_line = rmse_by_line.nlargest(15)
                
            plt.barh(range(len(rmse_by_line)), rmse_by_line.values)
            plt.yticks(range(len(rmse_by_line)), rmse_by_line.index)
            plt.title('Average RMSE by Line Item (Top 15)')
            plt.xlabel('RMSE')
        else:
            plt.text(0.5, 0.5, 'No valid RMSE data available', ha='center')
            plt.title('Average RMSE by Line Item')
        
        plt.tight_layout()
        pdf.savefig(fig, bbox_inches='tight')
        plt.close()
        
        # Add a section for Model Performance Summary
        fig = plt.figure(figsize=(15, 10))
        plt.suptitle('Model Performance Summary', fontsize=16, y=0.95)
        
        # Create a table with overall metrics
        overall_mape = metrics_df['MAPE'].replace([np.inf, -np.inf], np.nan).mean()
        overall_mae = metrics_df['MAE'].mean()
        overall_rmse = metrics_df['RMSE'].mean()
        
        # Create text summary instead of table
        plt.text(0.5, 0.6, f"Overall MAPE: {overall_mape:.2f}%", ha='center', fontsize=14)
        plt.text(0.5, 0.5, f"Overall MAE: {overall_mae:.2f}", ha='center', fontsize=14)
        plt.text(0.5, 0.4, f"Overall RMSE: {overall_rmse:.2f}", ha='center', fontsize=14)
        
        # Add explanation
        explanation = (
            "MAPE: Mean Absolute Percentage Error - Lower is better\n"
            "MAE: Mean Absolute Error - Absolute difference between predictions and actuals\n"
            "RMSE: Root Mean Square Error - Penalizes larger errors more heavily"
        )
        plt.text(0.5, 0.2, explanation, ha='center', fontsize=10)
        
        plt.axis('off')
        pdf.savefig(fig, bbox_inches='tight')
        plt.close()
        
        print(f"Comprehensive report saved to {output_filename}")
        return output_filename

# Usage example:
# generate_comprehensive_report(comparison_df, metrics_df, 'time_series_analysis_report.pdf')
```

```python
import boto3
import pandas as pd
from io import StringIO

def read_s3_csv(bucket_name, file_key):
    """
    Read CSV file from S3 bucket and rename first column to Date
    
    Args:
        bucket_name (str): Name of the S3 bucket
        file_key (str): Path to file within the bucket
        
    Returns:
        pd.DataFrame: DataFrame containing the CSV data with renamed first column
    """
    try:
        # Create S3 client
        s3_client = boto3.client('s3')
        
        # Get the object from S3
        response = s3_client.get_object(Bucket=bucket_name, Key=file_key)
        
        # Read the CSV content
        csv_content = response['Body'].read().decode('utf-8')
        
        # Convert to DataFrame
        df = pd.read_csv(StringIO(csv_content))
        
        # Rename first column to Date
        df = df.rename(columns={df.columns[0]: 'Date'})
        
        print(f"Successfully read file from S3")
        print(f"Shape of DataFrame: {df.shape}")
        print("\nFirst few rows:")
        print(df.head())
        print("\nColumns:")
        print(df.columns.tolist())
        
        return df
        
    except Exception as e:
        print(f"Error reading file from S3: {str(e)}")
        return None

# Define the S3 bucket and file path
bucket_name = "forecastingdataset08112024"
file_key = "10_02_2025_BS_IS_fred_complete/future_forecasts_2027 (4).csv"

# Read the data
future_forecasts = read_s3_csv(bucket_name, file_key)

# Additional data exploration if needed
if future_forecasts is not None:
    print("\nDataset Info:")
    print(future_forecasts.info())
    
    print("\nMissing Values:")
    print(future_forecasts.isnull().sum())
```

```python
import pandas as pd
import numpy as np
import json
from datetime import datetime

def create_extended_forecast(predictor, full_train_data, future_econ_factors, key_value_dict):
    """
    Create extended forecasts using trained model and future economic factors
    
    Modified to handle the specific format of unique_id values and to properly 
    extract components without triggering "Columns must be same length as key" error.
    
    Parameters:
    -----------
    predictor : SageMaker predictor object
        The trained DeepAR model predictor
    
    full_train_data : list
        The training data in JSON format
    
    future_econ_factors : pandas DataFrame
        DataFrame containing future economic factors with 'Date' column
    
    key_value_dict : dict
        Mapping from category ID to unique_id
        
    Returns:
    --------
    future_predictions : pandas DataFrame
        DataFrame containing the predictions with all identifiers
    """
    # Configure prediction
    predictor_input = {
        "instances": full_train_data,
        "configuration": {
            "output_types": ["mean", "quantiles"], 
            "quantiles": ["0.1", "0.5", "0.9"],
            "num_samples": 100
        }
    }
    
    # Get predictions
    predictions = predictor.predict(predictor_input)
    
    # Process predictions
    def process_predictions(response, econ_df):
        data = json.loads(response)
        dataframes = []
        
        # Get prediction length from the first prediction's mean array
        pred_length = len(data["predictions"][0]["mean"])
        
        # Prepare economic factors data - remove duplicates if any
        temp_df = econ_df.copy()
        temp_df['Date'] = pd.to_datetime(temp_df['Date'])
        temp_df = temp_df.drop_duplicates(subset=['Date'], keep='first')
        
        # Use future economic data date range
        start_date = temp_df['Date'].min()
        forecast_dates = pd.date_range(start=start_date, periods=pred_length, freq='MS')
        
        for idx, ts in enumerate(data["predictions"]):
            df = pd.DataFrame({
                'MonthData': forecast_dates,
                'mean': ts.get('mean', []),
                'Predicted_10': ts.get('quantiles', {}).get('0.1', []),
                'Predicted_50': ts.get('quantiles', {}).get('0.5', []),
                'Predicted_90': ts.get('quantiles', {}).get('0.9', []),
                'cat': ts.get('cat', idx)
            })
            
            # Merge with future economic factors
            df = pd.merge(df, 
                         temp_df,
                         left_on='MonthData', 
                         right_on='Date', 
                         how='left',
                         validate='1:1')  # Ensure one-to-one merge
            
            df = df.drop('Date', axis=1)
            dataframes.append(df)
        
        return pd.concat(dataframes, ignore_index=True)
    
    # Process and format predictions
    future_predictions = process_predictions(predictions, future_econ_factors)
    
    # Add unique_id mapping
    future_predictions['unique_id'] = future_predictions['cat'].map(key_value_dict)
    
    # Add separate identifier columns by splitting the unique_id
    # Check the format of the first unique_id to determine the splitting approach
    # This handles both formats (with ** or with _)
    sample_id = future_predictions['unique_id'].iloc[0] if not future_predictions.empty else ""
    
    try:
        # Try to extract identifier components based on the detected format
        if "**" in sample_id:
            # For format like "DataSourceId**ReportType**LineEntityId**BankName**LineEntityId_name**BankId"
            # Extract individual components separately to avoid the "columns must be same length as key" error
            future_predictions['DataSourceId'] = future_predictions['unique_id'].str.split('**', expand=True)[0]
            future_predictions['ReportType'] = future_predictions['unique_id'].str.split('**', expand=True)[1]
            future_predictions['LineEntityId'] = future_predictions['unique_id'].str.split('**', expand=True)[2]
            future_predictions['BankName'] = future_predictions['unique_id'].str.split('**', expand=True)[3]
            future_predictions['LineEntityId_name'] = future_predictions['unique_id'].str.split('**', expand=True)[4]
            future_predictions['BankId'] = future_predictions['unique_id'].str.split('**', expand=True)[5]
        else:
            # For format like "DataSourceId_BankId_LineEntityId"
            future_predictions[['DataSourceId', 'BankId', 'LineEntityId']] = future_predictions['unique_id'].str.split('_', expand=True)
    except Exception as e:
        print(f"Warning: Error splitting unique_id into components: {e}")
        # Fallback extraction for the underscore format (safer approach)
        try:
            future_predictions['DataSourceId'] = future_predictions['unique_id'].str.extract(r'^(\d+)_')
            future_predictions['BankId'] = future_predictions['unique_id'].str.extract(r'_(\d+)_')
            future_predictions['LineEntityId'] = future_predictions['unique_id'].str.extract(r'_(\d+)$')
        except Exception as e2:
            print(f"Warning: Fallback extraction also failed: {e2}")
            # Even simpler fallback - just leave the unique_id as is
            print("Using unique_id without splitting into components")
    
    return future_predictions

# 1. Prepare economic factors
future_econ_factors = future_forecasts.copy()
future_econ_factors['Date'] = pd.to_datetime(future_econ_factors['Date'])
future_econ_factors = future_econ_factors.drop_duplicates(subset=['Date'], keep='first')

print("Current future_forecasts range:", 
      future_econ_factors['Date'].min().strftime('%Y-%m-%d'), 
      "to", 
      future_econ_factors['Date'].max().strftime('%Y-%m-%d'))


numeric_columns = ['Prime', 'Discount', 'Discount PCR', 'Fed Funds', 'CPI', 'PCE', 'PPI', 
                  'GDP', 'Real GDP', 'Unemployment', 'Housing Starts', 'Housing Sales', 
                  'Consumer Sentiment', 'Treasury Spreads', '10 year Treas', '2 year Treas']

for col in numeric_columns:
    future_econ_factors[col] = pd.to_numeric(future_econ_factors[col], errors='coerce')

# 3. Create full training data
full_df = pd.concat([train_df, test_df]).sort_values(['unique_id', 'MonthData'])
full_train_data = create_json_format(full_df)

# 4. Generate extended forecasts
extended_forecasts = create_extended_forecast(predictor, full_train_data, future_econ_factors, key_value_dict)

# 5. Save results
extended_forecasts.to_csv('extended_forecasts_2027.csv', index=False)
print("Extended forecasts have been generated and saved.")
```

```python
def generate_forecast_visualizations(future_forecasts, output_base_name='forecast_2027', value_column='Prime'):
    """
    Generate detailed visualization PDFs for forecasts with proper column handling
    
    Args:
        future_forecasts: DataFrame containing forecast data
        output_base_name: Base name for output PDF files
        value_column: Column to use for forecast values (default: 'Prime')
    """
    # First clean up duplicate rows and map columns
    df = future_forecasts.copy()
    
    # Map columns to match expected format
    if 'Date' in df.columns and 'MonthData' not in df.columns:
        df['MonthData'] = pd.to_datetime(df['Date'])
    
    # Ensure proper datetime format
    df['MonthData'] = pd.to_datetime(df['MonthData'])
    df[value_column] = pd.to_numeric(df[value_column], errors='coerce')
    
    # Remove duplicates
    df = df.drop_duplicates()
    
    def create_forecast_plot(data, title, value_col=value_column):
        plt.figure(figsize=(15, 7))
        plt.plot(data['MonthData'], data[value_col], 
                label=f'{value_col} Forecast', color='blue', linewidth=2)
        
        plt.title(title, fontsize=12, pad=20)
        plt.xlabel('Date', fontsize=10)
        plt.ylabel(value_col, fontsize=10)
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()

    # 1. Overall Summary PDF
    with PdfPages(f'{output_base_name}_summary.pdf') as pdf:
        # Title page
        plt.figure(figsize=(12, 8))
        plt.text(0.5, 0.5, 'Forecast Analysis Summary\n' + 
                f'Generated on {pd.Timestamp.now().strftime("%Y-%m-%d")}',
                ha='center', va='center', fontsize=16)
        plt.axis('off')
        pdf.savefig()
        plt.close()

        # Summary statistics
        plt.figure(figsize=(12, 8))
        summary_text = (
            f"Date Range: {df['MonthData'].min().strftime('%Y-%m-%d')} to {df['MonthData'].max().strftime('%Y-%m-%d')}\n"
            f"Number of Time Points: {len(df['MonthData'].unique())}\n"
            f"Average {value_column}: {df[value_column].mean():,.2f}\n"
            f"{value_column} Range: {df[value_column].min():,.2f} to {df[value_column].max():,.2f}"
        )
        plt.text(0.5, 0.5, summary_text, ha='center', va='center', fontsize=12)
        plt.axis('off')
        pdf.savefig()
        plt.close()

        # Overall trend plot
        plt.figure(figsize=(15, 7))
        mean_by_date = df.groupby('MonthData')[value_column].mean()
        plt.plot(mean_by_date.index, mean_by_date.values, 
                label=f'Average {value_column}', linewidth=2)
        plt.title(f'Overall {value_column} Forecast Trend', fontsize=12, pad=20)
        plt.xlabel('Date', fontsize=10)
        plt.ylabel(value_column, fontsize=10)
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        pdf.savefig()
        plt.close()

    # Generate separate plots for each economic indicator
    numeric_columns = ['Prime', 'Discount', 'Discount PCR', 'Fed Funds', 'CPI', 'PCE', 'PPI', 
                      'GDP', 'Real GDP', 'Unemployment', 'Housing Starts', 'Housing Sales', 
                      'Consumer Sentiment', 'Treasury Spreads', '10 year Treas', '2 year Treas']
    
    # 2. Detailed Analysis PDF
    with PdfPages(f'{output_base_name}_detailed.pdf') as pdf:
        plt.figure(figsize=(12, 8))
        plt.text(0.5, 0.5, 'Detailed Economic Indicators Analysis\n' + 
                f'Generated on {pd.Timestamp.now().strftime("%Y-%m-%d")}',
                ha='center', va='center', fontsize=16)
        plt.axis('off')
        pdf.savefig()
        plt.close()

        # Plot each economic indicator
        for indicator in numeric_columns:
            if indicator in df.columns:
                df[indicator] = pd.to_numeric(df[indicator], errors='coerce')
                
                # Indicator summary
                plt.figure(figsize=(12, 8))
                indicator_summary = (
                    f"Indicator: {indicator}\n"
                    f"Average Value: {df[indicator].mean():,.2f}\n"
                    f"Range: {df[indicator].min():,.2f} to {df[indicator].max():,.2f}"
                )
                plt.text(0.5, 0.5, indicator_summary, ha='center', va='center', fontsize=12)
                plt.axis('off')
                pdf.savefig()
                plt.close()
                
                # Indicator forecast plot
                plt.figure(figsize=(15, 7))
                plt.plot(df['MonthData'], df[indicator], 
                        label=indicator, linewidth=2)
                plt.title(f"{indicator} Forecast", fontsize=12, pad=20)
                plt.xlabel('Date', fontsize=10)
                plt.ylabel('Value', fontsize=10)
                plt.xticks(rotation=45)
                plt.legend()
                plt.grid(True, alpha=0.3)
                plt.tight_layout()
                pdf.savefig()
                plt.close()

    print(f"Generated visualization PDFs: \n"
          f"1. {output_base_name}_summary.pdf\n"
          f"2. {output_base_name}_detailed.pdf")
```

```python
def generate_complete_trend_plots(historical_df, forecast_df, output_prefix='complete_forecast'):
    """
    Generate visualizations showing both historical and forecasted economic indicators
    """
    # Ensure proper datetime format
    historical_df = historical_df.copy()
    forecast_df = forecast_df.copy()
    
    # Standardize date columns
    historical_df['MonthData'] = pd.to_datetime(historical_df['MonthData'])
    forecast_df['MonthData'] = pd.to_datetime(forecast_df['Date'])
    
    # Define economic indicators
    economic_indicators = [
        'Prime', 'Discount', 'Discount PCR', 'Fed Funds', 'CPI', 'PCE', 'PPI', 
        'GDP', 'Real GDP', 'Unemployment', 'Housing Starts', 'Housing Sales', 
        'Consumer Sentiment', 'Treasury Spreads', '10 year Treas', '2 year Treas'
    ]
    
    def create_indicator_plot(data, indicator, title):
        """Helper function to create plots for economic indicators"""
        plt.figure(figsize=(20, 10))
        
        # Convert to numeric and plot
        data[indicator] = pd.to_numeric(data[indicator], errors='coerce')
        plt.plot(data['MonthData'], data[indicator], 
                label=indicator, color='blue', linewidth=2)
        
        # Add trend line
        z = np.polyfit(range(len(data)), data[indicator].fillna(method='ffill'), 1)
        p = np.poly1d(z)
        plt.plot(data['MonthData'], p(range(len(data))), 
                label='Trend', color='red', linestyle='--', alpha=0.5)
        
        plt.title(title, fontsize=14, pad=20)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Value', fontsize=12)
        plt.xticks(rotation=45)
        plt.legend(fontsize=10)
        plt.grid(True, alpha=0.3)
        
        # Add statistics box
        stats_text = (
            f"Statistics:\n"
            f"Mean: {data[indicator].mean():.2f}\n"
            f"Min: {data[indicator].min():.2f}\n"
            f"Max: {data[indicator].max():.2f}\n"
            f"Std Dev: {data[indicator].std():.2f}"
        )
        plt.text(0.02, 0.98, stats_text, transform=plt.gca().transAxes,
                verticalalignment='top', fontsize=10,
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        return plt.gcf()

    # 1. Individual Indicator Plots
    with PdfPages(f'{output_prefix}_indicators.pdf') as pdf:
        # Title page
        plt.figure(figsize=(12, 8))
        plt.text(0.5, 0.6, 'Economic Indicators Analysis\n' + 
                f'(2024-2027)\n\nGenerated on {pd.Timestamp.now().strftime("%Y-%m-%d")}',
                ha='center', va='center', fontsize=16)
        plt.text(0.5, 0.4, f'Total Indicators: {len(economic_indicators)}',
                ha='center', va='center', fontsize=14)
        plt.axis('off')
        pdf.savefig()
        plt.close()

        # Plot each indicator
        for indicator in economic_indicators:
            if indicator in forecast_df.columns:
                # Convert indicator values to numeric
                forecast_df[indicator] = pd.to_numeric(forecast_df[indicator], errors='coerce')
                
                # Summary page for indicator
                plt.figure(figsize=(12, 8))
                current_val = float(forecast_df[indicator].iloc[-1])
                starting_val = float(forecast_df[indicator].iloc[0])
                
                summary_text = (
                    f"Indicator: {indicator}\n\n"
                    f"Date Range: {forecast_df['MonthData'].min().strftime('%Y-%m-%d')} to "
                    f"{forecast_df['MonthData'].max().strftime('%Y-%m-%d')}\n\n"
                    f"Current Value: {current_val:.2f}\n"
                    f"Starting Value: {starting_val:.2f}\n"
                    f"Change: {(current_val - starting_val):.2f}"
                )
                plt.text(0.5, 0.5, summary_text, ha='center', va='center', fontsize=12)
                plt.axis('off')
                pdf.savefig()
                plt.close()
                
                # Trend plot
                fig = create_indicator_plot(
                    forecast_df,
                    indicator,
                    f"Forecast Trend for {indicator}"
                )
                pdf.savefig(fig)
                plt.close()

    # 2. Correlation Analysis
    with PdfPages(f'{output_prefix}_correlations.pdf') as pdf:
        plt.figure(figsize=(12, 8))
        plt.text(0.5, 0.6, 'Economic Indicators Correlation Analysis\n' + 
                f'Generated on {pd.Timestamp.now().strftime("%Y-%m-%d")}',
                ha='center', va='center', fontsize=16)
        plt.axis('off')
        pdf.savefig()
        plt.close()

        # Calculate and plot correlation matrix
        corr_data = forecast_df[economic_indicators].apply(pd.to_numeric, errors='coerce')
        correlation_matrix = corr_data.corr()
        
        plt.figure(figsize=(15, 12))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
                   fmt='.2f', square=True)
        plt.title('Correlation Matrix of Economic Indicators', pad=20)
        plt.tight_layout()
        pdf.savefig()
        plt.close()

    print(f"Generated visualization PDFs: \n"
          f"1. {output_prefix}_indicators.pdf - Individual indicator trends\n"
          f"2. {output_prefix}_correlations.pdf - Correlation analysis")

# Use the function
generate_complete_trend_plots(
    historical_df=train_df,
    forecast_df=future_forecasts,
    output_prefix='economic_forecast'
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def generate_comprehensive_report(comparison_df, metrics_df, output_filename='comprehensive_analysis.pdf', 
                                  limit_banks=False, limit_lines=False):
    """
    Generate a single comprehensive PDF report containing all time series analyses
    
    Parameters:
    -----------
    comparison_df : pandas DataFrame
        DataFrame containing actual vs predicted values
        
    metrics_df : pandas DataFrame
        DataFrame containing performance metrics (MAPE, MAE, RMSE)
        
    output_filename : str
        Filename for the output PDF report
        
    limit_banks : bool, default=False
        If True, only analyze top 15 banks by volume
        
    limit_lines : bool, default=False
        If True, only analyze top 10 line items by volume
    """
    # Convert datatypes and ensure consistency
    comparison_df['MonthData'] = pd.to_datetime(comparison_df['MonthData'])
    comparison_df['Actual'] = pd.to_numeric(comparison_df['Actual'], errors='coerce')
    comparison_df['Predicted_Mean'] = pd.to_numeric(comparison_df['Predicted_Mean'], errors='coerce')
    
    # Determine which column structure is available
    has_descriptive_columns = all(col in comparison_df.columns for col in ['BankName', 'LineEntityId_name'])
    
    # Define bank and line item columns based on available structure
    if has_descriptive_columns:
        bank_col = 'BankName'
        line_col = 'LineEntityId_name'
    else:
        bank_col = 'BankId'
        line_col = 'LineEntityId'
    
    with PdfPages(output_filename) as pdf:
        # Title page
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.7, 'Comprehensive Time Series Analysis Report', 
                ha='center', va='center', fontsize=24, fontweight='bold')
        plt.text(0.5, 0.5, f'Generated on {pd.Timestamp.now().strftime("%Y-%m-%d %H:%M:%S")}',
                ha='center', va='center', fontsize=14)
        summary_stats = (
            f"Total Banks: {len(comparison_df[bank_col].unique())}\n"
            f"Total Line Items: {len(comparison_df[line_col].unique())}\n"
            f"Date Range: {comparison_df['MonthData'].min().strftime('%Y-%m-%d')} to {comparison_df['MonthData'].max().strftime('%Y-%m-%d')}"
        )
        plt.text(0.5, 0.3, summary_stats, ha='center', va='center', fontsize=12)
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        # Section 1: Individual Bank Analysis
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.5, 'Section 1: Individual Bank Analysis', 
                ha='center', va='center', fontsize=20, fontweight='bold')
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        # Get the full list of banks or limit if requested
        bank_list = sorted(comparison_df[bank_col].unique())
        if limit_banks and len(bank_list) > 15:
            print(f"Limiting analysis to top 15 banks out of {len(bank_list)}")
            # Get top banks by transaction volume
            volume_by_bank = comparison_df.groupby(bank_col)['Actual'].sum().abs()
            bank_list = volume_by_bank.nlargest(15).index.tolist()
        else:
            print(f"Analyzing all {len(bank_list)} banks")
        
        for bank in bank_list:
            print(f"Processing bank: {bank}")
            df_subset = comparison_df[comparison_df[bank_col] == bank].copy()
            
            # Bank summary plot
            fig = plt.figure(figsize=(15, 10))
            plt.suptitle(f'Analysis for {bank_col}: {bank}', fontsize=16, y=0.95)
            
            # Time series plot
            plt.subplot(2, 1, 1)
            
            # Get all line items or limit depending on parameter
            line_items = df_subset[line_col].unique()
            if limit_lines and len(line_items) > 7:
                # Get top line items by volume
                volume_by_line = df_subset.groupby(line_col)['Actual'].sum().abs()
                line_items = volume_by_line.nlargest(7).index.tolist()
                plt.title(f'Time Series by Top 7 Line Items (out of {len(df_subset[line_col].unique())})')
            else:
                plt.title('Time Series by Line Item')
            
            # Create a colormap for distinguishing lines
            num_lines = len(line_items)
            colors = plt.cm.tab20(np.linspace(0, 1, num_lines))
            
            for i, line_item in enumerate(line_items):
                line_data = df_subset[df_subset[line_col] == line_item]
                plt.plot(line_data['MonthData'], line_data['Actual'], 
                        label=f'{line_item} (Actual)', 
                        linestyle='-', marker='o', color=colors[i])
                plt.plot(line_data['MonthData'], line_data['Predicted_Mean'], 
                        label=f'{line_item} (Predicted)', 
                        linestyle='--', marker='x', color=colors[i])
            
            plt.xticks(rotation=45)
            
            # Handle legend placement for many items
            if len(line_items) > 10:
                plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
            else:
                plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
                
            plt.grid(True, alpha=0.3)
            
            # Error analysis
            plt.subplot(2, 1, 2)
            errors = df_subset['Predicted_Mean'] - df_subset['Actual']
            plt.bar(df_subset['MonthData'], errors, alpha=0.6)
            plt.title('Prediction Errors')
            plt.xticks(rotation=45)
            plt.grid(True, alpha=0.3)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close()
        
        # Section 2: Line Item Analysis
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.5, 'Section 2: Line Item Analysis', 
                ha='center', va='center', fontsize=20, fontweight='bold')
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        # Get all line items or limit if requested
        line_list = sorted(comparison_df[line_col].unique())
        if limit_lines and len(line_list) > 10:
            print(f"Limiting analysis to top 10 line items out of {len(line_list)}")
            # Get top line items by transaction volume
            volume_by_line = comparison_df.groupby(line_col)['Actual'].sum().abs()
            line_list = volume_by_line.nlargest(10).index.tolist()
        else:
            print(f"Analyzing all {len(line_list)} line items")
        
        for line_item in line_list:
            print(f"Processing line item: {line_item}")
            df_subset = comparison_df[comparison_df[line_col] == line_item].copy()
            
            # Line item summary plot
            fig = plt.figure(figsize=(15, 10))
            plt.suptitle(f'Analysis for {line_col}: {line_item}', fontsize=16, y=0.95)
            
            # Box plot of actuals vs predictions
            plt.subplot(2, 1, 1)
            data_to_plot = [df_subset['Actual'].dropna(), df_subset['Predicted_Mean'].dropna()]
            plt.boxplot(data_to_plot, labels=['Actual', 'Predicted'])
            plt.title('Distribution of Values')
            plt.grid(True, alpha=0.3)
            
            # Time series by bank
            plt.subplot(2, 1, 2)
            
            # Get all banks or limit based on parameter
            banks = df_subset[bank_col].unique()
            if limit_banks and len(banks) > 7:
                # Get top banks by volume for this line item
                volume_by_bank = df_subset.groupby(bank_col)['Actual'].sum().abs()
                banks = volume_by_bank.nlargest(7).index.tolist()
                plt.title(f'Time Series by Top 7 Banks (out of {len(df_subset[bank_col].unique())})')
            else:
                plt.title('Time Series by Bank')
             
            # Create a colormap for distinguishing banks
            num_banks = len(banks)
            bank_colors = plt.cm.tab20(np.linspace(0, 1, num_banks))
                
            for i, bank in enumerate(banks):
                bank_data = df_subset[df_subset[bank_col] == bank]
                plt.plot(bank_data['MonthData'], bank_data['Actual'], 
                        label=f'{bank} (Actual)', marker='o', 
                        color=bank_colors[i % len(bank_colors)])
            
            plt.xticks(rotation=45)
            
            # Handle legend placement for many banks
            if len(banks) > 10:
                plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
            else:
                plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
                
            plt.grid(True, alpha=0.3)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close()
        
        # Overall Performance Metrics
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.5, 'Section 3: Overall Performance Metrics', 
                ha='center', va='center', fontsize=20, fontweight='bold')
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        # Ensure metrics_df has the right columns
        has_metrics_descriptive = all(col in metrics_df.columns for col in ['BankName', 'LineEntityId_name'])
        
        if has_metrics_descriptive != has_descriptive_columns:
            # If comparison_df and metrics_df have different column structures,
            # we need to adjust metrics_df to use the same bank and line columns
            metrics_bank_col = 'BankName' if has_metrics_descriptive else 'BankId'
            metrics_line_col = 'LineEntityId_name' if has_metrics_descriptive else 'LineEntityId'
        else:
            metrics_bank_col = bank_col
            metrics_line_col = line_col
        
        # Overall metrics plot
        fig = plt.figure(figsize=(15, 10))
        plt.suptitle('Overall Performance Metrics', fontsize=16, y=0.95)
        
        # MAPE by bank - with error handling for infinite/NaN values
        plt.subplot(2, 1, 1)
        
        if metrics_bank_col in metrics_df.columns:
            mape_by_bank = metrics_df.groupby(metrics_bank_col)['MAPE'].mean()
            mape_by_bank = mape_by_bank.replace([np.inf, -np.inf], np.nan).dropna().sort_values()
            
            if not mape_by_bank.empty:
                # Show all banks or limit based on parameter
                if limit_banks and len(mape_by_bank) > 15:
                    mape_by_bank = mape_by_bank.nlargest(15)
                    plt.title(f'Average MAPE by {metrics_bank_col} (Top 15)')
                else:
                    plt.title(f'Average MAPE by {metrics_bank_col}')
                
                # Use smaller font for many banks
                fontsize = 8 if len(mape_by_bank) > 20 else 10
                    
                plt.barh(range(len(mape_by_bank)), mape_by_bank.values)
                plt.yticks(range(len(mape_by_bank)), mape_by_bank.index, fontsize=fontsize)
                plt.xlabel('MAPE (%)')
            else:
                plt.text(0.5, 0.5, 'No valid MAPE data available', ha='center')
                plt.title(f'Average MAPE by {metrics_bank_col}')
        else:
            plt.text(0.5, 0.5, f'Column {metrics_bank_col} not found in metrics data', ha='center')
            plt.title('Average MAPE by Bank')
        
        # RMSE by line item
        plt.subplot(2, 1, 2)
        
        if metrics_line_col in metrics_df.columns:
            rmse_by_line = metrics_df.groupby(metrics_line_col)['RMSE'].mean().sort_values()
            
            if not rmse_by_line.empty:
                # Show all line items or limit based on parameter
                if limit_lines and len(rmse_by_line) > 15:
                    rmse_by_line = rmse_by_line.nlargest(15)
                    plt.title(f'Average RMSE by {metrics_line_col} (Top 15)')
                else:
                    plt.title(f'Average RMSE by {metrics_line_col}')
                
                # Use smaller font for many line items
                fontsize = 8 if len(rmse_by_line) > 20 else 10
                    
                plt.barh(range(len(rmse_by_line)), rmse_by_line.values)
                plt.yticks(range(len(rmse_by_line)), rmse_by_line.index, fontsize=fontsize)
                plt.xlabel('RMSE')
            else:
                plt.text(0.5, 0.5, 'No valid RMSE data available', ha='center')
                plt.title(f'Average RMSE by {metrics_line_col}')
        else:
            plt.text(0.5, 0.5, f'Column {metrics_line_col} not found in metrics data', ha='center')
            plt.title('Average RMSE by Line Item')
        
        plt.tight_layout()
        pdf.savefig(fig, bbox_inches='tight')
        plt.close()
        
        # Add overall metrics summary page
        fig = plt.figure(figsize=(15, 10))
        plt.suptitle('Overall Model Performance Summary', fontsize=16, y=0.95)
        
        # Calculate overall metrics
        overall_mape = metrics_df['MAPE'].replace([np.inf, -np.inf], np.nan).mean()
        overall_mae = metrics_df['MAE'].mean()
        overall_rmse = metrics_df['RMSE'].mean()
        
        # Create summary text
        summary_text = (
            f"Overall MAPE: {overall_mape:.2f}%\n\n"
            f"Overall MAE: {overall_mae:.2f}\n\n"
            f"Overall RMSE: {overall_rmse:.2f}\n\n"
            f"Number of Series Analyzed: {len(metrics_df)}\n\n"
            f"Date Range: {comparison_df['MonthData'].min().strftime('%Y-%m-%d')} to {comparison_df['MonthData'].max().strftime('%Y-%m-%d')}"
        )
        
        plt.text(0.5, 0.5, summary_text, ha='center', va='center', fontsize=14)
        
        # Add explanation of metrics
        explanation = (
            "MAPE: Mean Absolute Percentage Error - Lower is better\n"
            "MAE: Mean Absolute Error - Average absolute difference between predictions and actuals\n"
            "RMSE: Root Mean Square Error - Penalizes larger errors more heavily"
        )
        plt.text(0.5, 0.2, explanation, ha='center', fontsize=10)
        
        plt.axis('off')
        pdf.savefig(fig, bbox_inches='tight')
        plt.close()
        
        print(f"Comprehensive report saved to {output_filename}")
        return output_filename

# Example usage:
# generate_comprehensive_report(comparison_df, metrics_df, 'financial_time_series_report.pdf', limit_banks=False, limit_lines=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns

def combine_historical_future(historical_df, future_df):
    """
    Combine historical and future data with flexible column handling
    
    This function will adapt to whatever columns are available in both dataframes
    and fill in missing columns with appropriate default values.
    """
    # Ensure date columns are datetime
    historical_df['MonthData'] = pd.to_datetime(historical_df['MonthData'])
    future_df['MonthData'] = pd.to_datetime(future_df['MonthData'])
    
    # Check which columns are available in future_df
    future_columns = {
        'MonthData': future_df['MonthData'],
        'unique_id': future_df['unique_id'],
        'Actual': np.nan,  # No actuals for future dates
        'Predicted_Mean': future_df['mean'] if 'mean' in future_df.columns else future_df['Predicted_Mean']
    }
    
    # Add quantile predictions if available
    for col in ['Predicted_10', 'Predicted_50', 'Predicted_90']:
        if col in future_df.columns:
            future_columns[col] = future_df[col]
    
    # Add identifier columns if available
    id_columns = ['DataSourceId', 'BankId', 'LineEntityId', 'ReportType', 'BankName', 'LineEntityId_name']
    for col in id_columns:
        if col in future_df.columns:
            future_columns[col] = future_df[col]
        elif col in historical_df.columns:
            # If column exists in historical but not future, we'll try to recreate it
            if col == 'DataSourceId' and 'unique_id' in future_df.columns:
                # Extract DataSourceId from unique_id (assuming format: DataSourceId_BankId_LineEntityId)
                future_columns[col] = future_df['unique_id'].str.extract(r'^(\d+)_').iloc[:, 0]
            elif col == 'BankId' and 'unique_id' in future_df.columns:
                # Extract BankId from unique_id
                future_columns[col] = future_df['unique_id'].str.extract(r'_(\d+)_').iloc[:, 0]
            elif col == 'LineEntityId' and 'unique_id' in future_df.columns:
                # Extract LineEntityId from unique_id
                future_columns[col] = future_df['unique_id'].str.extract(r'_(\d+)$').iloc[:, 0]
            else:
                # Create an empty column for other identifiers
                future_columns[col] = np.nan
    
    # Create the formatted future dataframe
    future_df_formatted = pd.DataFrame(future_columns)
    
    # Ensure all columns from historical_df exist in future_df_formatted
    for col in historical_df.columns:
        if col not in future_df_formatted.columns:
            future_df_formatted[col] = np.nan
    
    # Ensure all columns from future_df_formatted exist in historical_df
    historical_df_copy = historical_df.copy()
    for col in future_df_formatted.columns:
        if col not in historical_df.columns:
            historical_df_copy[col] = np.nan
    
    # Combine and sort
    combined_df = pd.concat([historical_df_copy, future_df_formatted], ignore_index=True)
    return combined_df.sort_values(['unique_id', 'MonthData'])

def generate_comprehensive_report(combined_df, output_filename='comprehensive_forecast_analysis.pdf', 
                                 limit_banks=False, limit_lines=False):
    """
    Generate a detailed PDF report with historical and forecast analysis
    
    Parameters:
    -----------
    combined_df : pandas DataFrame
        DataFrame containing both historical and future forecast data
        
    output_filename : str
        Filename for the output PDF report
        
    limit_banks : bool, default=False
        If True, only analyze top 15 banks by volume
        
    limit_lines : bool, default=False
        If True, only analyze top 10 line items by volume
    """
    # Ensure data types are consistent
    combined_df['MonthData'] = pd.to_datetime(combined_df['MonthData'])
    combined_df['Actual'] = pd.to_numeric(combined_df['Actual'], errors='coerce')
    combined_df['Predicted_Mean'] = pd.to_numeric(combined_df['Predicted_Mean'], errors='coerce')
    
    # Check for quantile columns and convert if they exist
    for col in ['Predicted_10', 'Predicted_50', 'Predicted_90']:
        if col in combined_df.columns:
            combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')
    
    # Determine which columns to use for grouping
    # First check if descriptive columns exist
    has_bank_name = 'BankName' in combined_df.columns and combined_df['BankName'].notna().any()
    has_line_name = 'LineEntityId_name' in combined_df.columns and combined_df['LineEntityId_name'].notna().any()
    
    # Define bank and line item columns based on available data
    if has_bank_name:
        bank_col = 'BankName'
    elif 'BankId' in combined_df.columns:
        bank_col = 'BankId'
    else:
        # Extract bank from unique_id if it exists
        if 'unique_id' in combined_df.columns:
            combined_df['BankId'] = combined_df['unique_id'].str.extract(r'_(\d+)_').iloc[:, 0]
            bank_col = 'BankId'
        else:
            raise ValueError("No bank identifier column found")
    
    if has_line_name:
        line_col = 'LineEntityId_name'
    elif 'LineEntityId' in combined_df.columns:
        line_col = 'LineEntityId'
    else:
        # Extract line item from unique_id if it exists
        if 'unique_id' in combined_df.columns:
            combined_df['LineEntityId'] = combined_df['unique_id'].str.extract(r'_(\d+)$').iloc[:, 0]
            line_col = 'LineEntityId'
        else:
            raise ValueError("No line item identifier column found")
    
    with PdfPages(output_filename) as pdf:
        # Title page
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.7, 'Comprehensive Forecast Analysis Report', 
                ha='center', va='center', fontsize=24, fontweight='bold')
        plt.text(0.5, 0.5, f'Generated on {pd.Timestamp.now().strftime("%Y-%m-%d %H:%M:%S")}',
                ha='center', va='center', fontsize=14)
        
        summary_stats = (
            f"Total Banks: {len(combined_df[bank_col].unique())}\n"
            f"Total Line Items: {len(combined_df[line_col].unique())}\n"
            f"Date Range: {combined_df['MonthData'].min().strftime('%Y-%m-%d')} to "
            f"{combined_df['MonthData'].max().strftime('%Y-%m-%d')}"
        )
        plt.text(0.5, 0.3, summary_stats, ha='center', va='center', fontsize=12)
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        # Section header for Bank Analysis
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.5, 'Section 1: Individual Bank Analysis', 
                ha='center', va='center', fontsize=20, fontweight='bold')
        plt.axis('off')
        pdf.savefig()
        plt.close()

        # Get all banks or limit if requested
        bank_list = sorted(combined_df[bank_col].unique())
        
        # Determine which banks to analyze
        if limit_banks and len(bank_list) > 15:
            print(f"Limiting analysis to top 15 banks out of {len(bank_list)}")
            # Get top banks by transaction volume (using absolute values to account for negative amounts)
            historical_data = combined_df[~combined_df['Actual'].isna()]
            volume_by_bank = historical_data.groupby(bank_col)['Actual'].sum().abs()
            bank_list = volume_by_bank.nlargest(15).index.tolist()
        else:
            print(f"Analyzing all {len(bank_list)} banks")

        # Bank Analysis Section
        for bank in bank_list:
            print(f"Processing bank: {bank}")
            df_bank = combined_df[combined_df[bank_col] == bank].copy()
            
            # Bank summary plot
            fig = plt.figure(figsize=(15, 10))
            plt.suptitle(f'Analysis for {bank_col}: {bank}', fontsize=16, y=0.95)
            
            # Get all line items or limit based on parameter
            line_items = df_bank[line_col].unique()
            
            if limit_lines and len(line_items) > 7:
                # Get top line items by volume
                historical_line_data = df_bank[~df_bank['Actual'].isna()]
                if not historical_line_data.empty:
                    volume_by_line = historical_line_data.groupby(line_col)['Actual'].sum().abs()
                    line_items = volume_by_line.nlargest(7).index.tolist()
                plt.title(f'Time Series by Top 7 Line Items (out of {len(df_bank[line_col].unique())})')
            else:
                plt.title('Time Series by Line Item')
            
            # Time series by line item
            plt.subplot(2, 1, 1)
            
            # Create a colormap for distinguishing lines
            num_lines = len(line_items)
            colors = plt.cm.tab20(np.linspace(0, 1, num_lines))
            
            # Identify historical/forecast boundary
            try:
                historical_end = combined_df[~combined_df['Actual'].isna()]['MonthData'].max()
                plt.axvline(x=historical_end, color='gray', linestyle='--', alpha=0.7)
                plt.text(historical_end, 0, ' Historical | Forecast ', 
                        backgroundcolor='w', ha='center')
            except:
                # In case there's no historical data
                pass
            
            for i, line_item in enumerate(line_items):
                line_data = df_bank[df_bank[line_col] == line_item]
                
                # Plot historical actuals if they exist
                historical_mask = ~line_data['Actual'].isna()
                if any(historical_mask):
                    plt.plot(line_data.loc[historical_mask, 'MonthData'], 
                            line_data.loc[historical_mask, 'Actual'],
                            label=f'{line_item} (Historical)', linestyle='-', marker='o', 
                            color=colors[i % len(colors)])
                
                # Plot predictions (both historical and future)
                predict_mask = ~line_data['Predicted_Mean'].isna()
                if any(predict_mask):
                    plt.plot(line_data.loc[predict_mask, 'MonthData'], 
                            line_data.loc[predict_mask, 'Predicted_Mean'],
                            label=f'{line_item} (Forecast)', linestyle='--', alpha=0.7,
                            color=colors[i % len(colors)])
                
                # Add confidence intervals if available
                if all(col in line_data.columns for col in ['Predicted_10', 'Predicted_90']):
                    valid_mask = ~line_data['Predicted_10'].isna() & ~line_data['Predicted_90'].isna()
                    if any(valid_mask):
                        plt.fill_between(line_data.loc[valid_mask, 'MonthData'],
                                      line_data.loc[valid_mask, 'Predicted_10'],
                                      line_data.loc[valid_mask, 'Predicted_90'],
                                      alpha=0.1, color=colors[i % len(colors)])
            
            plt.xticks(rotation=45)
            
            # Handle legend placement for many items
            if len(line_items) > 10:
                plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
            else:
                plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
                
            plt.grid(True, alpha=0.3)
            
            # Forecast accuracy metrics
            plt.subplot(2, 1, 2)
            historical_data = df_bank[~df_bank['Actual'].isna() & ~df_bank['Predicted_Mean'].isna()]
            
            if not historical_data.empty:
                errors = historical_data['Predicted_Mean'] - historical_data['Actual']
                plt.hist(errors, bins=min(30, len(errors)), alpha=0.6)
                plt.title('Historical Prediction Error Distribution')
                plt.xlabel('Prediction Error')
                plt.ylabel('Frequency')
                
                # Add error statistics
                try:
                    mae = np.mean(np.abs(errors))
                    
                    # Calculate MAPE only for non-zero actual values to avoid div/0
                    non_zero_mask = historical_data['Actual'] != 0
                    if any(non_zero_mask):
                        valid_errors = errors[non_zero_mask]
                        valid_actuals = historical_data.loc[non_zero_mask, 'Actual']
                        mape = np.mean(np.abs(valid_errors / valid_actuals)) * 100
                    else:
                        mape = np.nan
                        
                    rmse = np.sqrt(np.mean(errors**2))
                    
                    stats_text = f"MAE: {mae:.2f}\nMAPE: {mape:.2f}%\nRMSE: {rmse:.2f}"
                    plt.text(0.95, 0.95, stats_text, transform=plt.gca().transAxes, 
                            va='top', ha='right', bbox=dict(facecolor='white', alpha=0.7))
                except:
                    # In case of error in statistics calculation
                    pass
            else:
                plt.text(0.5, 0.5, 'No historical data available for error analysis', 
                        ha='center', va='center')
                plt.title('Historical Prediction Error Distribution')
                
            plt.grid(True, alpha=0.3)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close()

        # Section header for Line Item Analysis
        plt.figure(figsize=(15, 10))
        plt.text(0.5, 0.5, 'Section 2: Line Item Analysis', 
                ha='center', va='center', fontsize=20, fontweight='bold')
        plt.axis('off')
        pdf.savefig()
        plt.close()
        
        # Get all line items or limit if requested
        line_list = sorted(combined_df[line_col].unique())
        
        if limit_lines and len(line_list) > 10:
            print(f"Limiting analysis to top 10 line items out of {len(line_list)}")
            # Get top line items by transaction volume
            historical_data = combined_df[~combined_df['Actual'].isna()]
            if not historical_data.empty:
                volume_by_line = historical_data.groupby(line_col)['Actual'].sum().abs()
                line_list = volume_by_line.nlargest(10).index.tolist()
        else:
            print(f"Analyzing all {len(line_list)} line items")

        # Line Item Analysis Section
        for line_item in line_list:
            print(f"Processing line item: {line_item}")
            df_line = combined_df[combined_df[line_col] == line_item].copy()
            
            fig = plt.figure(figsize=(15, 10))
            plt.suptitle(f'Analysis for {line_col}: {line_item}', fontsize=16, y=0.95)
            
            # Get all banks or limit based on parameter
            banks = df_line[bank_col].unique()
            
            if limit_banks and len(banks) > 7:
                # Get top banks by volume for this line item
                historical_data = df_line[~df_line['Actual'].isna()]
                if not historical_data.empty:
                    volume_by_bank = historical_data.groupby(bank_col)['Actual'].sum().abs()
                    banks = volume_by_bank.nlargest(7).index.tolist()
                plt.title(f'Time Series by Top 7 Banks (out of {len(df_line[bank_col].unique())})')
            else:
                plt.title('Time Series by Bank')
            
            # Time series by bank
            plt.subplot(2, 1, 1)
            
            # Create a colormap for distinguishing banks
            num_banks = len(banks)
            bank_colors = plt.cm.tab20(np.linspace(0, 1, num_banks))
            
            # Identify historical/forecast boundary
            try:
                historical_end = combined_df[~combined_df['Actual'].isna()]['MonthData'].max()
                plt.axvline(x=historical_end, color='gray', linestyle='--', alpha=0.7)
                plt.text(historical_end, 0, ' Historical | Forecast ', 
                        backgroundcolor='w', ha='center')
            except:
                # In case there's no historical data
                pass
            
            for i, bank in enumerate(banks):
                bank_data = df_line[df_line[bank_col] == bank]
                
                # Plot historical actuals
                historical_mask = ~bank_data['Actual'].isna()
                if any(historical_mask):
                    plt.plot(bank_data.loc[historical_mask, 'MonthData'],
                            bank_data.loc[historical_mask, 'Actual'],
                            label=f'{bank} (Historical)', marker='o', alpha=0.7,
                            color=bank_colors[i % len(bank_colors)])
                
                # Plot future predictions
                future_mask = bank_data['Actual'].isna() & ~bank_data['Predicted_Mean'].isna()
                if any(future_mask):
                    plt.plot(bank_data.loc[future_mask, 'MonthData'],
                            bank_data.loc[future_mask, 'Predicted_Mean'],
                            label=f'{bank} (Forecast)', linestyle='--', alpha=0.7,
                            color=bank_colors[i % len(bank_colors)])
            
            plt.xticks(rotation=45)
            
            # Handle legend placement for many banks
            if len(banks) > 10:
                plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
            else:
                plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
                
            plt.grid(True, alpha=0.3)
            
            # Forecast distribution
            plt.subplot(2, 1, 2)
            future_data = df_line[df_line['Actual'].isna() & ~df_line['Predicted_Mean'].isna()]
            
            if not future_data.empty and len(future_data) > 1:
                # Check if we have all quantile columns
                has_quantiles = all(col in future_data.columns for col in ['Predicted_10', 'Predicted_90'])
                
                if has_quantiles:
                    plt.boxplot([future_data['Predicted_10'], future_data['Predicted_Mean'], 
                                future_data['Predicted_90']], 
                               labels=['10th Percentile', 'Mean', '90th Percentile'])
                else:
                    # Just show the mean forecast
                    plt.boxplot([future_data['Predicted_Mean']], labels=['Mean Forecast'])
                    
                plt.title('Distribution of Future Forecasts')
                
                # Add forecast growth statistics
                try:
                    first_forecast = future_data.iloc[0]['Predicted_Mean']
                    last_forecast = future_data.iloc[-1]['Predicted_Mean']
                    
                    if first_forecast != 0 and not np.isnan(first_forecast) and not np.isnan(last_forecast):
                        total_growth = (last_forecast / first_forecast - 1) * 100
                        annualized_growth = ((last_forecast / first_forecast) ** (12 / len(future_data)) - 1) * 100
                        
                        growth_text = (f"Total growth over forecast period: {total_growth:.2f}%\n"
                                      f"Annualized growth rate: {annualized_growth:.2f}%")
                        
                        plt.text(0.95, 0.95, growth_text, transform=plt.gca().transAxes, 
                                va='top', ha='right', bbox=dict(facecolor='white', alpha=0.7))
                except:
                    # In case of error in growth calculation
                    pass
            else:
                plt.text(0.5, 0.5, 'Insufficient future forecast data available', 
                        ha='center', va='center')
                plt.title('Distribution of Future Forecasts')
                
            plt.grid(True, alpha=0.3)
            
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight')
            plt.close()
            
        # Add overall forecast summary page
        plt.figure(figsize=(15, 10))
        plt.suptitle('Overall Forecast Summary', fontsize=16, y=0.95)
        
        # Calculate historical vs forecast comparison
        historical_data = combined_df[~combined_df['Actual'].isna()]
        future_data = combined_df[combined_df['Actual'].isna() & ~combined_df['Predicted_Mean'].isna()]
        
        # Create summary text
        try:
            historical_end = historical_data['MonthData'].max() if not historical_data.empty else None
            forecast_start = future_data['MonthData'].min() if not future_data.empty else None
            forecast_end = future_data['MonthData'].max() if not future_data.empty else None
            
            if historical_end and forecast_start and forecast_end:
                forecast_months = (forecast_end.year - forecast_start.year) * 12 + forecast_end.month - forecast_start.month + 1
                
                summary_text = (
                    f"Historical Period: Up to {historical_end.strftime('%Y-%m-%d')}\n\n"
                    f"Forecast Period: {forecast_start.strftime('%Y-%m-%d')} to {forecast_end.strftime('%Y-%m-%d')}\n"
                    f"({forecast_months} months)\n\n"
                    f"Total Banks Analyzed: {len(combined_df[bank_col].unique())}\n\n"
                    f"Total Line Items Analyzed: {len(combined_df[line_col].unique())}\n\n"
                )
            else:
                # Handle case where we might not have both historical and forecast data
                summary_text = (
                    f"Date Range: {combined_df['MonthData'].min().strftime('%Y-%m-%d')} to "
                    f"{combined_df['MonthData'].max().strftime('%Y-%m-%d')}\n\n"
                    f"Total Banks Analyzed: {len(combined_df[bank_col].unique())}\n\n"
                    f"Total Line Items Analyzed: {len(combined_df[line_col].unique())}\n\n"
                )
        except:
            # Fallback summary if there's an error
            summary_text = "Analysis completed successfully."
        
        # Add overall growth statistics if possible
        try:
            if not historical_data.empty and not future_data.empty:
                agg_historical = historical_data.groupby('MonthData')['Actual'].sum()
                agg_forecast = future_data.groupby('MonthData')['Predicted_Mean'].sum()
                
                if not agg_historical.empty and not agg_forecast.empty:
                    last_historical = agg_historical.iloc[-1]
                    first_forecast = agg_forecast.iloc[0]
                    last_forecast = agg_forecast.iloc[-1]
                    
                    if last_historical != 0 and first_forecast != 0:
                        initial_change = (first_forecast / last_historical - 1) * 100
                        total_growth = (last_forecast / first_forecast - 1) * 100
                        
                        growth_text = (
                            f"Initial forecast change from last historical point: {initial_change:.2f}%\n\n"
                            f"Total growth over forecast period: {total_growth:.2f}%\n\n"
                            f"Annualized growth rate: {((last_forecast / first_forecast) ** (12 / len(agg_forecast)) - 1) * 100:.2f}%"
                        )
                        
                        summary_text += growth_text
        except:
            summary_text += "Unable to calculate overall growth statistics."
            
        plt.text(0.5, 0.5, summary_text, ha='center', va='center', fontsize=14)
        plt.axis('off')
        pdf.savefig()
        plt.close()

        print(f"Comprehensive report saved to {output_filename}")
        return output_filename

def main():
    try:
        # Load historical data
        print("Loading historical data...")
        historical_df = pd.read_csv('actuals_vs_predicted_detailed_202405_to_202410.csv')
        print(f"Historical data loaded: {historical_df.shape} rows, columns: {list(historical_df.columns)}")
        
        # Load future predictions with correct filename
        print("Loading future predictions...")
        future_df = pd.read_csv('extended_forecasts_2027.csv')
        print(f"Future data loaded: {future_df.shape} rows, columns: {list(future_df.columns)}")
        
        if 'Date' in future_df.columns:
            future_df = future_df.rename(columns={'Date': 'MonthData'})
        
        # Combine datasets
        print("Combining datasets...")
        combined_df = combine_historical_future(historical_df, future_df)
        print(f"Combined data: {combined_df.shape} rows, columns: {list(combined_df.columns)}")
        
        # Generate comprehensive report with ALL banks and line items
        print("Generating comprehensive report...")
        generate_comprehensive_report(combined_df, 'comprehensive_forecast_analysis_2027.pdf', 
                                    limit_banks=False, limit_lines=False)
        
        # Save combined dataset
        combined_df.to_csv('combined_historical_future_data_2027.csv', index=False)
        print("Combined dataset saved to combined_historical_future_data_2027.csv")
        
    except Exception as e:
        print(f"Error in main function: {str(e)}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np

def calculate_monthly_mape(df):
    """
    Calculate MAPE for each month and unique_id combination where actuals exist
    
    Args:
        df: DataFrame with columns ['MonthData', 'unique_id', 'Actual', 'Predicted_Mean']
    
    Returns:
        DataFrame with MAPE calculations added
    """
    # Create a copy of the input DataFrame
    result_df = df.copy()
    
    # Convert MonthData to datetime if it's not already
    result_df['MonthData'] = pd.to_datetime(result_df['MonthData'])
    
    # Convert Actual and Predicted_Mean to numeric, handling any non-numeric values
    result_df['Actual'] = pd.to_numeric(result_df['Actual'], errors='coerce')
    result_df['Predicted_Mean'] = pd.to_numeric(result_df['Predicted_Mean'], errors='coerce')
    
    # Calculate MAPE only where we have both actual and predicted values
    mask = ~(result_df['Actual'].isna() | result_df['Predicted_Mean'].isna())
    result_df['MAPE'] = np.nan  # Initialize MAPE column with NaN
    
    # Calculate MAPE where we have both values
    result_df.loc[mask, 'MAPE'] = (
        np.abs(result_df.loc[mask, 'Actual'] - result_df.loc[mask, 'Predicted_Mean']) / 
        np.abs(result_df.loc[mask, 'Actual']) * 100
    )
    
    # Add summary statistics for validation
    validation_stats = {
        'total_rows': len(result_df),
        'rows_with_mape': mask.sum(),
        'unique_ids': len(result_df['unique_id'].unique()),
        'date_range': f"{result_df['MonthData'].min()} to {result_df['MonthData'].max()}"
    }
    
    print("\nValidation Statistics:")
    for key, value in validation_stats.items():
        print(f"{key}: {value}")
        
    # Calculate and print average MAPE by month
    monthly_mape = result_df.groupby(result_df['MonthData'].dt.strftime('%Y-%m'))['MAPE'].mean()
    print("\nAverage MAPE by Month:")
    print(monthly_mape)
    
    return result_df

def update_combined_data(input_file, output_file):
    """
    Read combined data file, calculate MAPE, and save updated version
    
    Args:
        input_file: Path to input CSV file
        output_file: Path to save updated CSV file
    """
    # Read the combined data
    df = pd.read_csv(input_file)
    
    # Calculate MAPE
    updated_df = calculate_monthly_mape(df)
    
    # Save updated data
    updated_df.to_csv(output_file, index=False)
    print(f"\nUpdated data saved to {output_file}")
    
    # Return sample of data for verification
    print("\nSample of updated data (first 5 rows):")
    sample_cols = ['MonthData', 'unique_id', 'Actual', 'Predicted_Mean', 'MAPE']
    print(updated_df[sample_cols].head())
    
    return updated_df

# Example usage:
updated_df = update_combined_data('combined_historical_future_data_2027.csv', 
                                'combined_historical_future_data_2027_with_mape.csv')
```

```python
import pandas as pd
import numpy as np

def add_forecast_metrics(input_file, output_file):
    """
    Calculate MAPE, MAE, and RMSE for each month and unique_id combination 
    for the period 2024-05 to 2024-10
    
    Args:
        input_file (str): Path to input CSV file
        output_file (str): Path to output CSV file
    
    Returns:
        pd.DataFrame: Updated DataFrame with forecast accuracy metrics
    """
    # Read the data
    df = pd.read_csv(input_file)
    
    # Convert dates and numeric columns
    df['MonthData'] = pd.to_datetime(df['MonthData'])
    df['Actual'] = pd.to_numeric(df['Actual'], errors='coerce')
    df['Predicted_Mean'] = pd.to_numeric(df['Predicted_Mean'], errors='coerce')
    
    # Initialize metrics columns
    df['MAPE'] = np.nan
    df['MAE'] = np.nan
    df['RMSE'] = np.nan
    
    # Calculate metrics only where we have both actual and predicted values
    mask = ~(df['Actual'].isna() | df['Predicted_Mean'].isna())
    
    # Calculate MAPE
    df.loc[mask, 'MAPE'] = (
        np.abs(df.loc[mask, 'Actual'] - df.loc[mask, 'Predicted_Mean']) / 
        np.abs(df.loc[mask, 'Actual']) * 100
    )
    
    # Calculate MAE
    df.loc[mask, 'MAE'] = np.abs(df.loc[mask, 'Actual'] - df.loc[mask, 'Predicted_Mean'])
    
    # Calculate RMSE
    # Note: We calculate this at the individual level first, then take the sqrt at the group level
    df.loc[mask, 'RMSE'] = (df.loc[mask, 'Actual'] - df.loc[mask, 'Predicted_Mean']) ** 2
    
    # Filter validation period
    validation_period = (
        (df['MonthData'] >= '2024-05-01') & 
        (df['MonthData'] <= '2024-10-31')
    )
    validation_df = df[validation_period]
    
    # Calculate monthly statistics
    monthly_stats = validation_df.groupby(
        validation_df['MonthData'].dt.strftime('%Y-%m')
    ).agg({
        'MAPE': ['mean', 'std', 'min', 'max'],
        'MAE': ['mean', 'std', 'min', 'max'],
        'RMSE': lambda x: np.sqrt(np.mean(x))  # Proper RMSE calculation
    }).round(2)
    
    print("Monthly Forecast Metrics:")
    print("\nMonthly Statistics:")
    print(monthly_stats)
    
    # Calculate statistics by unique_id
    id_stats = validation_df.groupby('unique_id').agg({
        'MAPE': ['mean', 'std'],
        'MAE': ['mean', 'std'],
        'RMSE': lambda x: np.sqrt(np.mean(x))
    }).round(2)
    
    print("\nValidation Period Statistics:")
    print(f"Total rows in validation period: {len(validation_df)}")
    print(f"Rows with metrics calculated: {validation_df['MAPE'].notna().sum()}")
    print(f"Unique IDs in validation period: {validation_df['unique_id'].nunique()}")
    print("\nOverall Metrics:")
    print(f"Average MAPE: {validation_df['MAPE'].mean():.2f}%")
    print(f"Average MAE: {validation_df['MAE'].mean():.2f}")
    print(f"Overall RMSE: {np.sqrt(validation_df['RMSE'].mean()):.2f}")
    
    # Add aggregated metrics back to the dataframe
    # Calculate monthly RMSE for each unique_id
    monthly_rmse = df[mask].groupby(['unique_id', df['MonthData'].dt.strftime('%Y-%m')])['RMSE'].apply(
        lambda x: np.sqrt(np.mean(x))
    ).reset_index()
    monthly_rmse.columns = ['unique_id', 'month', 'RMSE']
    
    # Merge back the proper RMSE values
    df.loc[mask, 'RMSE'] = np.sqrt(df.loc[mask, 'RMSE'])
    
    # Save updated DataFrame
    df.to_csv(output_file, index=False)
    print(f"\nUpdated data saved to {output_file}")
    
    # Save detailed metrics by unique_id
    metrics_file = output_file.replace('.csv', '_metrics_by_id.csv')
    id_stats.to_csv(metrics_file)
    print(f"Detailed metrics by ID saved to {metrics_file}")
    
    return df, monthly_stats, id_stats

# Additional utility function to analyze metrics by specific dimensions
def analyze_metrics_by_dimension(df, dimension_col, metrics_cols=['MAPE', 'MAE', 'RMSE']):
    """
    Analyze forecast metrics by a specific dimension (e.g., BankName, LineEntityId_name)
    
    Args:
        df: DataFrame with metrics
        dimension_col: Column to group by
        metrics_cols: List of metrics to analyze
    
    Returns:
        DataFrame with metrics analysis by dimension
    """
    metrics_analysis = df.groupby(dimension_col).agg({
        metric: ['mean', 'std', 'min', 'max'] for metric in metrics_cols
    }).round(2)
    
    return metrics_analysis

# Usage example:
df, monthly_stats, id_stats = add_forecast_metrics(
    'combined_historical_future_data_2027.csv',
    'combined_historical_future_data_2027_with_metrics.csv'
)

# Optional: Analyze by specific dimensions
bank_analysis = analyze_metrics_by_dimension(df, 'BankName')
line_item_analysis = analyze_metrics_by_dimension(df, 'LineEntityId_name')
```

```python
import boto3
import pandas as pd
from io import StringIO

def read_s3_csv(bucket_name, file_key):
    """
    Read CSV file from S3 bucket
    
    Args:
        bucket_name (str): Name of the S3 bucket
        file_key (str): Path to file within the bucket
        
    Returns:
        pd.DataFrame: DataFrame containing the CSV data
    """
    try:
        # Create S3 client
        s3_client = boto3.client('s3')
        
        # Get the object from S3
        response = s3_client.get_object(Bucket=bucket_name, Key=file_key)
        
        # Read the CSV content
        csv_content = response['Body'].read().decode('utf-8')
        
        # Convert to DataFrame
        df = pd.read_csv(StringIO(csv_content))
        
        print(f"Successfully read file from S3")
        print(f"Shape of DataFrame: {df.shape}")
        print("\nFirst few rows:")
        print(df.head())
        print("\nColumns:")
        print(df.columns.tolist())
        
        # Basic data info
        print("\nDataset Info:")
        print(df.info())
        
        print("\nMissing Values:")
        print(df.isnull().sum())
        
        return df
        
    except Exception as e:
        print(f"Error reading file from S3: {str(e)}")
        return None

# Define the S3 bucket and file path
bucket_name = "forecastingdataset08112024"
file_key = "Comparing_outputs for both with and without fred/BS_ALL_2027 (1).csv"

# Read the data
comparison_data = read_s3_csv(bucket_name, file_key)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def create_detailed_plots(data, output_file='detailed_forecast_plots.pdf'):
    """
    Create detailed plots for each unique ID showing actuals, predictions, and forecasts
    
    Args:
        data (pd.DataFrame): DataFrame containing the data with 'unique_id', 'MonthData', 
                           'Amount', 'mean', '0.1', '0.9' columns
        output_file (str): Name of output PDF file
    """
    # Convert MonthData to datetime if not already
    data['MonthData'] = pd.to_datetime(data['MonthData'])
    
    # Define time periods
    prediction_start = pd.Timestamp('2024-05-01')
    prediction_end = pd.Timestamp('2024-10-31')
    
    # Get unique IDs
    unique_ids = data['unique_id'].unique()
    
    with PdfPages(output_file) as pdf:
        for uid in unique_ids:
            # Get data for this unique ID
            df_sub = data[data['unique_id'] == uid].copy()
            
            # Split data into time periods
            historical = df_sub[df_sub['MonthData'] < prediction_start].copy()
            prediction = df_sub[
                (df_sub['MonthData'] >= prediction_start) & 
                (df_sub['MonthData'] <= prediction_end)
            ].copy()
            forecast = df_sub[df_sub['MonthData'] > prediction_end].copy()
            
            # Create figure
            fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), height_ratios=[2, 1])
            plt.subplots_adjust(hspace=0.3)
            
            # Main plot (Actuals, Predictions, and Forecasts)
            # Historical data (blue)
            if not historical.empty:
                ax1.plot(historical['MonthData'], historical['Amount'], 
                        'b-', label='Historical (2015-2024)', linewidth=1.5, marker='.', markersize=2)
            
            # Prediction period (green for actual, red for predicted)
            if not prediction.empty:
                ax1.plot(prediction['MonthData'], prediction['Amount'], 
                        'g-', label='Recent Actual', linewidth=1.5, marker='o', markersize=3)
                ax1.plot(prediction['MonthData'], prediction['mean'],
                        'r--', label='Recent Predicted', linewidth=1.5)
                
                # Confidence intervals for prediction period
                ax1.fill_between(prediction['MonthData'],
                               prediction['0.1'],
                               prediction['0.9'],
                               color='red', alpha=0.1, label='Recent 80% CI')
            
            # Forecast period (purple)
            if not forecast.empty:
                ax1.plot(forecast['MonthData'], forecast['mean'],
                        'purple', linestyle='--', label='Future Forecast', linewidth=1.5)
                
                # Confidence intervals for forecast period
                ax1.fill_between(forecast['MonthData'],
                               forecast['0.1'],
                               forecast['0.9'],
                               color='purple', alpha=0.1, label='Forecast 80% CI')
            
            # Format main plot
            title = f"Forecast Analysis without FRED Economic Indicators\n{uid}"
            ax1.set_title(title, fontsize=10, pad=20)
            ax1.grid(True, alpha=0.3, linestyle='--')
            ax1.legend(loc='center left', bbox_to_anchor=(1.01, 0.5))
            ax1.set_ylabel('Value', fontsize=9)
            ax1.tick_params(axis='x', rotation=45)
            
            # Use scientific notation for y-axis
            ax1.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
            
            # Growth Rate subplot
            # Calculate growth rates
            for df in [historical, prediction, forecast]:
                if not df.empty:
                    if 'Amount' in df.columns:
                        df['Growth'] = df['Amount'].pct_change(periods=12) * 100
                    elif 'mean' in df.columns:
                        df['Growth'] = df['mean'].pct_change(periods=12) * 100
            
            # Plot growth rates
            if not historical.empty:
                ax2.plot(historical['MonthData'], historical['Growth'],
                        'b-', label='Historical Growth', linewidth=1.5)
            
            if not forecast.empty:
                ax2.plot(forecast['MonthData'], forecast['Growth'],
                        'r--', label='Forecast Growth', linewidth=1.5)
            
            # Format growth rate subplot
            ax2.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
            ax2.set_title('Year-over-Year Growth Rate (%)', fontsize=10)
            ax2.grid(True, alpha=0.3, linestyle='--')
            ax2.legend(loc='center left', bbox_to_anchor=(1.01, 0.5))
            ax2.set_xlabel('Date', fontsize=9)
            ax2.set_ylabel('Growth Rate (%)', fontsize=9)
            ax2.tick_params(axis='x', rotation=45)
            
            # Calculate statistics
            stats = {
                'Historical Mean': historical['Amount'].mean() if not historical.empty else np.nan,
                'Forecast Mean': forecast['mean'].mean() if not forecast.empty else np.nan,
                'Historical Growth': historical['Growth'].mean() if not historical.empty else np.nan,
                'Forecast Growth': forecast['Growth'].mean() if not forecast.empty else np.nan
            }
            
            stats_text = "\n".join([
                f"{k}: {v:,.2f}" + ("%" if "Growth" in k else "")
                for k, v in stats.items() if not np.isnan(v)
            ])
            
            # Add stats text
            plt.figtext(1.15, 0.5, stats_text,
                       bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'),
                       fontsize=8, verticalalignment='center')
            
            # Save plot
            plt.tight_layout()
            pdf.savefig(fig, bbox_inches='tight', pad_inches=0.5)
            plt.close()
            
    print(f"\nDetailed plots saved to: {output_file}")
    print(f"Number of plots generated: {len(unique_ids)}")

# Use the function
create_detailed_plots(
    data=filtered_data,
    output_file='detailed_forecast_plots.pdf'
)
```
